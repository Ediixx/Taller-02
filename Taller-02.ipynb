```python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de la funcin
def f(x):
    return x**3 - 3*x**2 + x - 1

x = np.linspace(-2, 4, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y
```

```python
from scipy.optimize import newton

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.newton.html
x0 = 100
points = []
x_r = newton(
    f, x0, args=(points, True), maxiter=100
)  # Este no es mtodo de Newton, es secante!
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.scatter(x_r, f(x_r), color="red", zorder=5)  # Add the root to the plot
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
# Extract x values from points
x_points = [point[0] for point in points]

# Plot the x points
plt.figure(figsize=(8, 6))
plt.plot(x_points, "o-", label="x points")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of x points in points")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.show()
```

```python
from scipy.optimize import newton

# Definicin de la funcin
def f(x):
    return x**3 - 3*x**2 + x - 1

# Definicin de la derivada de la funcin
def f_prime(x):
    return 3*x**2 - 6*x + 1

# Valor inicial
x0 = -1

# Utilizando el Mtodo de Newton con scipy y obteniendo la informacin adicional
x_r, alg_info = newton(
    f,                      
    x0,                     
    fprime=f_prime,        
    maxiter=30,            
    disp=False,            
    full_output=True        
)

# Informacin adicional sobre el algoritmo
print("Informacin del algoritmo:\n", alg_info)

# Mostrar la raz encontrada
print("Raz encontrada:", x_r)

# Graficar la funcin y la raz encontrada
x = np.linspace(-2, 4, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.scatter(x_r, f(x_r), color="orange", zorder=5)  
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

```
